```
#importing Libraries
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```
lic_data = cv2.CascadeClassifier('/content/haarcascade_russian_plate_number.xml')
```

```
def plt_show(image, title="", gray = False, size =(100,100)):
    temp = image
    if gray == False:
        temp = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
        plt.title(title)
        plt.imshow(temp, cmap='gray')
        plt.show()
```

```
def detect_number(img):
    temp = img
    gray = cv2.cvtColor(temp, cv2.COLOR_BGR2GRAY)
    number = lic_data.detectMultiScale(img,1.2)
    print("number plate detected:"+str(len(number)))
    for numbers in number:
        (x,y,w,h) = numbers
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+h]
        cv2.rectangle(temp, (x,y), (x+w,y+h), (0,255,0), 3)
        
    plt_show(temp)
```

```

img = cv2.imread("/content/car.png")
plt_show(img)
detect_number(img)
```
